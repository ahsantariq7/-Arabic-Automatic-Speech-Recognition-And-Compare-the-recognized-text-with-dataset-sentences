```python
import arabic_reshaper
from bidi.algorithm import get_display
import speech_recognition as sr
```

```python
# Preprocess the dataset
dataset_sentences = [
    "مرحبًا كيف يمكنني مساعدتك؟",
    "ما هو اسمك؟",
    "كيف حالك اليوم؟",
    "متى يمكننا اللقاء؟"
]

normalized_sentences = [get_display(arabic_reshaper.reshape(sentence)) for sentence in dataset_sentences]
```

```python
# Implement Automatic Speech Recognition (ASR)
def recognize_speech(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio = recognizer.record(source)
    recognized_text = recognizer.recognize_google(audio, language='ar')
    return recognized_text
```

```python
# Compare the recognized text with dataset sentences
def compare_text(recognized_text, dataset_sentences):
    mismatched_words = []
    matched_words = []

    recognized_words = recognized_text.split()
    for i, word in enumerate(recognized_words):
        if word in dataset_sentences[i]:
            matched_words.append((word, dataset_sentences[i]))
        else:
            mismatched_words.append((word, dataset_sentences[i]))

    return mismatched_words, matched_words
```

```python
# Example usage
audio_file_path = 'helo.wav'
recognized_text = recognize_speech(audio_file_path)
mismatched_words, matched_words = compare_text(recognized_text, normalized_sentences)
```

```python
print("Mismatched words:")
for word, dataset_sentence in mismatched_words:
    print(f"Word: {word}, Expected Sentence: {dataset_sentence}")

print("\nMatched words:")
for word, dataset_sentence in matched_words:
    print(f"Word: {word}, Matched Sentence: {dataset_sentence}")
```

```python
# Compare the recognized text with dataset sentences
def get_matched_sentence(recognized_text, dataset_sentences):
    recognized_words = recognized_text.split()
    for i, word in enumerate(recognized_words):
        if word in dataset_sentences[i]:
            return dataset_sentences[i]
    return None

# Example usage
audio_file_path = 'helo.wav'
recognized_text = recognize_speech(audio_file_path)
matched_sentence = get_matched_sentence(recognized_text, normalized_sentences)

# Print the result
if matched_sentence:
    print("Matched sentence:", matched_sentence)
else:
    print("No matching sentence found.")
```
